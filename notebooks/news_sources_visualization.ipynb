# Display All News Sources in the United States in English

```python
import requests
import pandas as pd
from great_tables import GT

url_sources = "https://newsapi.org/v2/top-headlines/sources??country=us&language=en&apiKey=206782d873e3487e94f4fbf2823b566b"
response_sources = requests.get(url_sources)
json_data_sources = response_sources.json()

df_sources = pd.json_normalize(json_data_sources['sources'])

# Create a display table
gt_tbl_sources = GT(df_sources)

# Now, show the gt table
gt_tbl_sources
```
