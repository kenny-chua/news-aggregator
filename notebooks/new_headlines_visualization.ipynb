# Top Headlines in the United States in English Visualization

```python
import requests
import pandas as pd
from great_tables import GT

url_headlines = "https://newsapi.org/v2/top-headlines?country=us&language=en&apiKey=206782d873e3487e94f4fbf2823b566b"
response_headlines = requests.get(url_headlines)
json_data_headlines = response_headlines.json()

df_headlines = pd.json_normalize(json_data_headlines['articles'])
```

#### Column Names for Top Headlines Response

```python
column_names_headlines = df_headlines.columns.tolist()
print(column_names_headlines)
```

```python
# Create a display table
gt_tbl_headlines = GT(df_headlines)

# Now, show the gt table
gt_tbl_headlines
```
